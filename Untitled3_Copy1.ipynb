```python
def detect_faces(file,classes,activation_model,required_size=(640,480)):
    detected_faces = detector.detect_faces(file)
    boxes_list = []
    scores_list = []
    class_list = []
    features_list = []
    
    for face in detected_faces:
        x,y,w,h = face['box']
        boxes_list.append(face['box'])        
        scores_list.append(face['confidence'])
        x2,y2 = (x+w),(y+h)
        pixels = file[y:y2,x:x2]
        pixels = cv2.resize(pixels,required_size)
        new_pixels = Image.fromarray(pixels)
        #new_pixels = new_pixels.resize(required_size)
        new_pixels = np.array(new_pixels)
        new_pixels = np.expand_dims(new_pixels,axis=0)
    
        new_pixels = np.asarray(new_pixels)
        prediction = new_model.predict(new_pixels)
        layer_out = activation_model.predict(new_pixels/255.0)
        
        #print('shape of prediction is :',prediction.shape)
        
        for pred in prediction:
            max_val = np.amax(pred)
            #print('value of max_val is ;',max_val)
            
            if max_val >= 0.46:
                class_val = np.argmax(pred)
                print('class_val is :',class_val)
                class_list.append(classes[class_val])
                
            else:
                class_list.append('unknown')
        
        feature = layer_out
        
        #print('shape of feature.flatten is:',feature.flatten().shape)
        features_list.append(feature.flatten())
        
    boxes_list = np.array(boxes_list)
    scores_list = np.array(scores_list)
    features_list = np.array(features_list)
    class_list = np.array(class_list)
    
    #print('after class_list')
    detections = [Detection(bbox,score,class_name,feature) for bbox,score,class_name,feature in 
                      zip(boxes_list,scores_list,class_list,features_list)]
    tracker.predict()
    #print('in tracker predict')
    tracker.update(detections)
    #print('in tracker update')
    
    cmap = plt.get_cmap('tab20b')
    colors = [cmap(i)[:3] for i in np.linspace(0,1,20)]
    current_count = int(0)
        
    for track in tracker.tracks:
        if not track.is_confirmed() or track.time_since_update >1:
            continue
        
        #print('in tracker.tracks')    
        bbox = track.to_tlbr()
        
        class_name= track.get_class()
        color = colors[int(track.track_id) % len(colors)]
        color = [i * 255 for i in color]
        
        cv2.rectangle(file, (int(bbox[0]),int(bbox[1])), (int(bbox[2]),int(bbox[3])), color, 1)
        cv2.putText(file, class_name, (int(bbox[0]), int(bbox[1]-10)), 0, 0.75,(0, 0, 255), 1)
        
    cv2.imshow('camera',file)   
    out.write(file)

    return 0
    
      
```
